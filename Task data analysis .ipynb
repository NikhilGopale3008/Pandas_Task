```python
## import Libraries
import pandas as pd
```

```python
# Try To Join Both DataFram 
a=inner_join=pd.merge(left=data1, right=data2, on='Student name',how='inner',
                     suffixes=('_data1','_data2'))
a.head(5)
```

```python
# concatenate  both datasets

data=pd.concat([data1,data2])
data
```

```python
# rename columns

data.rename(columns={'Student name':'Student_name','Exam name':'Exam_name',
                     'Exam points':'Exam_points','Exam1':'Exam1','Exam2':'Exam2'},inplace=True)
```

```python
data.head()
```

```python
# find null values and replace it by 0

data.isnull().sum()

data.fillna(0)
```
